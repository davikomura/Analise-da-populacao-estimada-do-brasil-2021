# Analise de dados de uma base de dados de e-commerce no ano de 2015

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('global_superstore_2016_pt-br.xlsx')
df['Ano'] = df['Data do pedido'].dt.year
df['Mês'] = df['Data do pedido'].dt.month
df['Dia'] = df['Data do pedido'].dt.day
```

```python
df.columns
```

```python
df_2015 = df[df['Ano'] == 2015]
```

## Total de vendas, total de quantidade e total de lucro do ano

```python
df_analise = df_2015[['Ano', 'Vendas', 'Quantidade', 'Lucro']].round(2)
```

```python
df_total_ano = df_analise.groupby(['Ano']).sum()
df_total_ano
```

##### O total de vendas no ano de 2015 foi de 4,3 milhões aproximadamente, já a quantidade foi de 60622 e o lucro de aproximadamente 505 mil.

## Gráficos das vendas diárias

#### Fiz o grafico diário de cada mês de 2015.

```python
df_vendas = df_2015[['Mês', 'Dia', 'Vendas']].round(2)
```

```python
def grafico_diario(x):
    df_vendas_mes2015 = df_vendas[df_vendas['Mês'] == x]

    df_vendas_diarias_mes2015 = df_vendas_mes2015.groupby(['Mês', 'Dia']).sum()

    return sns.lineplot(data = df_vendas_diarias_mes2015, x = 'Dia', y = 'Vendas', color='green')
```

#### Aqui fiz um DataFrame apenas com os Países, o lucro, as vendas e a quantidade.

```python
df_analise_paises = df_2015[['País', 'Lucro', 'Vendas', 'Quantidade']].round(2)
```

```python
df_analise_paises_total = df_analise_paises.groupby(['País']).sum()
```

## Top 5 de países com maiores lucros

```python
df_analise_paises_maiores_lucros = pd.DataFrame(df_analise_paises_total).sort_values(by = 'Lucro', 
                                                                                     ascending=False)
df_analise_paises_maiores_lucros.head()
```

## Top 5 de países com menores lucros

```python
df_analise_paises_menores_lucros = pd.DataFrame(df_analise_paises_total).sort_values(by = 'Lucro')
df_analise_paises_menores_lucros.head()
```

## Segmento com maior lucro

```python
df_analise_segmento = df_2015[['Segmento', 'Lucro', 'Vendas', 'Quantidade']].round(2)
```

```python
df_analise_segmento_lucro = df_analise_segmento.groupby(['Segmento']).sum()
df_analise_segmento_maior_lucro = pd.DataFrame(df_analise_segmento_lucro).sort_values(by = 'Lucro', 
                                                                                      ascending=False)
df_analise_segmento_maior_lucro.head(1)
```

## Categoria com maior lucro

```python
df_analise_categoria = df_2015[['Categoria', 'Lucro', 'Vendas', 'Quantidade']].round(2)
```

```python
df_analise_categoria_lucro = df_analise_categoria.groupby(['Categoria']).sum()
df_analise_categoria_maior_lucro = pd.DataFrame(df_analise_categoria_lucro).sort_values(by = 'Lucro', 
                                                                                        ascending=False)
df_analise_categoria_maior_lucro.head(1)
```

## Categoria com menor lucro

```python
df_analise_categoria_maior_lucro = pd.DataFrame(df_analise_categoria_lucro).sort_values(by = 'Lucro')
df_analise_categoria_maior_lucro.head(1)
```

## Top 10 de produtos com maiores vendas

```python
df_analise_produto = df_2015[['Nome do produto', 'Vendas', 'Lucro', 'Quantidade']].round(2)
```

```python
df_analise_produto_vendas = df_analise_produto.groupby(['Nome do produto']).sum()
df_analise_produto_maiores_vendas = pd.DataFrame(df_analise_produto_vendas).sort_values(by = 'Vendas', 
                                                                                        ascending=False)
df_analise_produto_maiores_vendas.head(10)
```

## Top 10 de produtos com menores vendas

```python
df_analise_produto_vendas = df_analise_produto.groupby(['Nome do produto']).sum()
df_analise_produto_menores_vendas = pd.DataFrame(df_analise_produto_vendas).sort_values(by = 'Vendas')
df_analise_produto_menores_vendas.head(10)
```
